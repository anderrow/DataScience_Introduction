## Operation with arrays 

```python
list_1 = [1, 2, 3]
list_2 = [4, 5, 6]
```

```python
array_1 = np.array(list_1)
array_2 = np.array(list_2)
```

```python
list_1 * list_2
```

Como se puede ver python no es capaz de multiplicar listas

```python
array_1*array_2
```

No obstante numpy nos permite hacerlo de forma eficiente y rápida, si quisieramos hacerlo en python habría que hacer un bucle for:

```python
[num1*num2 for num1, num2 in zip(list_1,list_2)]
```

# Estadística

## Distribuciones

```python
array_normal = np.random.normal(1.75, 0.05, 1000) #Genera una distribución normal con media de 1,75, desviacion estandar de 5 cm 
                                          #(Donde entran el 68% de todos los valores) con una muestra n = 1000
print (array_normal)
```

## np.mean

```python
np.mean(array_normal) #Media
```

## np.median

```python
np.median(array_normal) #Nos da la mediana, el valor que se encuentra en la mitad, en este caso el número 500 ya que n= = 1000
```

## np.std

```python
np.std (array_normal) #Desviación standar
```
