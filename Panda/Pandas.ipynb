## df.where

```python
df.where(df['especialidad'] == 'Programación') #Filtro por columnas Na significa No available
```

```python
filtro1 = df['especialidad'] == 'Programación'
```

```python
filtro2 = df['visitas'] > 145000
```

```python
df.where(df['especialidad'] == 'Programación').dropna() #Para quitar NaN
```

```python
df.where(filtro1 & filtro2).dropna() #Podemos combinar filtros
```

```python
df.where(filtro1 & filtro2).dropna()['nombre'] #El nuevo objeto creado es un dataframe podemos trabajar con él
```

```python
df.where(filtro1 & filtro2).dropna().loc[:, 'nombre'] #Todas las filas y columna de nombre
```

A esto se le llama **Method Chaining** o encadenamiento de metódos, cuando encadenas uno tras otro para generar algo.
